```python
import pandas as pd 
import numpy as np
import plotly
import plotly.express as px
import plotly.offline as pyo
import os

colour_theme = plotly.colors.sequential.Sunsetdark_r
graph_path = './ploty_visuals'
os.makedirs(graph_path,exist_ok=True)


#%matplotlib inline
```

```python
og_data = pd.read_csv('myntra_products.csv')
print(og_data.shape)
og_data.info()
```

```python
max_categorical_level = og_data['tags'].str.count(',').max() + 1
df = og_data.copy()
df[[f'categorical_level_{i+1}' for i in range(max_categorical_level)]] = df['tags'].str.split(',', expand=True)
```

```python
print(df.duplicated().sum())
df.drop_duplicates(keep='first',inplace=True)
```

```python
print(df.isnull().sum())
df[df['categorical_level_2'].isnull()]
```

# Understanding the Data

```python
vtitle = 'Category Ratio'
vfigure = px.pie(data_frame=df,names='categorical_level_1',title=vtitle, color_discrete_sequence=colour_theme)
vtitle = vtitle.replace(' ','_')
pyo.plot(vfigure,filename=f'{graph_path}/{vtitle}.html',auto_open=False)
vfigure.show()
```

```python
vtitle='categorical_level_1'
vfigure=px.histogram(data_frame=df,x=vtitle,text_auto=True,color_discrete_sequence=colour_theme)
vtitle = vtitle.replace(' ','_')
pyo.plot(vfigure,filename=f'{graph_path}/{vtitle}.html',auto_open=False)
vfigure.show()
```

```python

vtitle='Relations  with rating and category'
vfigure=px.histogram(data_frame=df[~df['categorical_level_1'].isin( ['Not Specified','free-items'])],x='categorical_level_1',y='rating_count',text_auto='.2f',title=vtitle,color_discrete_sequence=colour_theme)
vtitle = vtitle.replace(' ','_')
pyo.plot(vfigure,filename=f'{graph_path}/{vtitle}.html',auto_open=False)
vfigure.show()
```

### Top Wear

```python

vtitle='Top 20 mens topwear Brands Based On Products'
top20_men_topwear_brands=df[(df['categorical_level_2']=='men-clothing')& (df['product_tag'].isin(['tshirts','shirts']))].groupby(by=['brand_name','product_tag'])['product_name'].count().nlargest(20)
top20_men_topwear_brands = top20_men_topwear_brands.reset_index().rename({'brand_name':'brands','product_name':'products'},axis='columns')
vfigure=px.histogram(data_frame=top20_men_topwear_brands,y='brands',x= 'products',color='product_tag',title=vtitle,color_discrete_sequence=colour_theme)
vtitle = vtitle.replace(' ','_')
pyo.plot(vfigure,filename=f'{graph_path}/{vtitle}.html',auto_open=False)
vfigure.show()
```

```python
vtitle='Top 20 mens topwear Brands Based On Products And Rating'
men_tshirt_brands_grp = df[(df['categorical_level_2']=='men-clothing')& (df['product_tag'].isin(['tshirts','shirts']))].groupby(by=['brand_name','product_tag'])[['product_name','rating','rating_count']].aggregate(products = pd.NamedAgg(column='product_name',aggfunc='count'),avg_rating=pd.NamedAgg(column='rating',aggfunc='mean'),rating_count = pd.NamedAgg(column='rating_count',aggfunc='sum')).sort_values(by=['rating_count','avg_rating'],ascending=False).reset_index()
vfigure=px.bar(data_frame=men_tshirt_brands_grp.head(30),x='brand_name',y='avg_rating',color='product_tag',hover_data=['products','rating_count'],text_auto='.2f',title=vtitle,color_discrete_sequence=colour_theme)
vtitle = vtitle.replace(' ','_')
pyo.plot(vfigure,filename=f'{graph_path}/{vtitle}.html',auto_open=False)
vfigure.show()
```

```python
size_dict = {}
def extract_size(row):
     product = row['product_tag']
     sizes = str(row['sizes']).split(',')

     if size_dict.get(product,False):
          pass
     else :
          size_dict[product] = {}

     for size in sizes:
          if len(str(size).split('/')) >= 2:
               size = str(size).split('/')
               size = size[1]

          if size_dict[product].get(size,False):
               size_dict[product][size] += 1
          else:
               size_dict[product][size] = 1

     return None
    
```

```python
temp_null = df[df['categorical_level_2']=='men-clothing'].apply(extract_size,axis=1)
size_dict.keys()
```

```python
mens_clothing_size_df = pd.DataFrame(size_dict)
mens_clothing_size_df = mens_clothing_size_df.reset_index(names='size')
```

```python
vtitle='TopWear Size Available in Mens Clothing'
vfigure=px.histogram(data_frame=mens_clothing_size_df, x='size',y=['shirts','tshirts','kurtas','sweatshirts'],text_auto=True,title=vtitle,color_discrete_sequence=colour_theme)
vtitle = vtitle.replace(' ','_')
pyo.plot(vfigure,filename=f'{graph_path}/{vtitle}.html',auto_open=False)
vfigure.show()
```
